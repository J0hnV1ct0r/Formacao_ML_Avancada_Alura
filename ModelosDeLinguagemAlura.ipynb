# Filtro 1

```python
def remover(textos, regex):
  if type(textos) == str:
    textos = regex.sub("", textos)
    return textos
  else:
    return [regex.sub("", texto) for texto in textos]
```

```python
#Teste da função e regex
regex_html = re.compile(r'<.*?>')
quest_sem_tag = remover(quest_port, regex_html)
print(quest_sem_tag)
```

```python
quest_sem_tag = remover(overflow_ingles.Questão[5], regex_html)
print(quest_sem_tag )
```

# Filtro 2

```python
def sub_codigo(textos, regex):
  if type(textos) == str:
    textos = regex.sub("CODE", textos)
    return textos
  else:
    return [regex.sub("CODE", texto) for texto in textos]
```

```python
regex_codigo = re.compile(r'<code>(.|(\n))*?</code>', flags=re.DOTALL)
```

```python
quest_sem_code = sub_codigo(quest_port, regex_codigo)
print(quest_sem_code)
```

```python
quest_sem_code = sub_codigo(overflow_ingles.Questão[5], regex_codigo)
print(quest_sem_code)
```
