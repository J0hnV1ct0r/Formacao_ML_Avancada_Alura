# Filtro 1

```python
def remover(textos, regex):
  if type(textos) == str:
    textos = regex.sub("", textos)
    return textos
  else:
    return [regex.sub("",texto) for texto in textos]
```

```python
#Teste da função e regex
regex_html = re.compile(r'<.*?>')
quest_sem_tag = remover(quest_port, regex_html)
print(quest_sem_tag)
```

```python
quest_sem_tag = remover(overflow_ingles.Questão[5], regex_html)
print(quest_sem_tag )
```

# Filtro 2

```python
def sub_codigo(textos, regex):
  if type(textos) == str:
    textos = regex.sub("CODE", textos)
    return textos
  else:
    return [regex.sub("CODE", texto) for texto in textos]
```

```python
regex_codigo = re.compile(r'<code>(.|(\n))*?</code>', flags=re.DOTALL)
```

```python
quest_sem_code = sub_codigo(quest_port, regex_codigo)
print(quest_sem_code)
```

```python
quest_sem_code = sub_codigo(overflow_ingles.Questão[5], regex_codigo)
print(quest_sem_code)
```

# Tratamento dos dados 1

```python
def tratamento_1(textos):
  regex_html = re.compile(r'<.*?>')
  regex_codigo = re.compile(r'<code>(.|(\n))*?</code>', flags=re.DOTALL)
  questões_port_sem_code = sub_codigo(textos, regex_codigo)#Filtro de substituir codigo
  questões_port_sem_code_tag = remover(questões_port_sem_code, regex_html)#Filtro de remover tags html
  return questões_port_sem_code_tag
```

```python
#Tratamento 1 é a aplicação dos filtros 1 e 2 em sequencia
overflow_port['Tratamento_sem_code_tag']=tratamento_1(overflow_port.Questão)
overflow_port.head(100)
```

```python
print(overflow_port.Tratamento_sem_code_tag[0])
```

```python
#Demonstração do tratamento com questões em ingles
questoes_ingles_sem_code = sub_codigo(overflow_ingles.Questão, regex_codigo)#Filtro de substituir codigo
questoes_ingles_sem_code_tag = remover(questoes_ingles_sem_code, regex_html)#Filtro de remover tags html
overflow_ingles['Tratamento_sem_code_tag']=questoes_ingles_sem_code_tag
overflow_ingles.head(100)
```

```python
print(overflow_ingles.Tratamento_sem_code_tag[0])
```

# Filtro 3

```python
def minusculo(textos):
  if type(textos) == str:
    textos = textos.lower()
    return textos
  else:
    return [texto.lower() for texto in textos]
```

```python
#remove a pontuação do texto
regex_pontuacao = re.compile(r'[^\w\s]')
quest_sem_pontuacao = remover(overflow_port.Tratamento_sem_code_tag[0], regex_pontuacao)
print(quest_sem_pontuacao)
```

```python
print(minusculo(quest_sem_pontuacao))
```

#Filtro 4

```python
#remove os digitos do texto
regex_digitos = re.compile(r'\d+')
quest_sem_digitos = remover(overflow_port.Tratamento_sem_code_tag[0], regex_digitos)
print(quest_sem_digitos)
```

# Filtro 5

```python
# Detecta espaçoes duplicados
regex_espacos = re.compile(r' +')
```

```python
def remover_espacos(textos, regex):
  if type(textos) == str:
    texto = regex.sub(" ",textos)
    return texto
  else:
    return [regex.sub(" ",texto) for texto in textos]
```

```python
print(remover_espacos(quest_sem_digitos, regex_espacos))
```

# Tratamento 2

```python
def tratamento_2(textos):
  regex_pontuacao = re.compile(r'[^\w\s]')
  regex_digitos = re.compile(r'\d+')
  regex_espacos = re.compile(r' +')

  questoes_port_sem_pont = sub_codigo(textos, regex_codigo)#Filtro de substituir codigo
  questoes_port_min = minusculo(questoes_port_sem_pont)#Filtro tornar minusculo
  questoes_port_sem_dig = remover(questoes_port_min, regex_digitos)#Filtro de remover digitos
  questoes_port_espaco = remover_espacos(questoes_port_sem_dig, regex_espacos)#Filtro de remover espaços duplicados
  return questoes_port_espaco
```

```python
#Tratamento 2 é a aplicação dos filtros 3, 4, 5 e de tornar minusculo em sequencia
overflow_port['Tratamento_final']=tratamento_2(overflow_port.Tratamento_sem_code_tag)
overflow_port.head(100)
```

```python
overflow_ingles['Tratamento_final']=tratamento_2(overflow_ingles.Tratamento_sem_code_tag)
overflow_ingles.head(100)
```
