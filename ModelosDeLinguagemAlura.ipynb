# Filtro 1

```python
def remover(textos, regex):
  if type(textos) == str:
    textos = regex.sub("", textos)
    return textos
  else:
    return [regex.sub("",texto) for texto in textos]
```

# Filtro 2

```python
def sub_codigo(textos, regex):
  if type(textos) == str:
    textos = regex.sub("CODE", textos)
    return textos
  else:
    return [regex.sub("CODE", texto) for texto in textos]
```

```python
regex_codigo = re.compile(r'<code>(.|(\n))*?</code>', flags=re.DOTALL)
```

```python
def minusculo(textos):
  if type(textos) == str:
    textos = textos.lower()
    return textos
  else:
    return [texto.lower() for texto in textos]
```

# Tratamento 2

```python
def tratamento_2(textos):
  regex_pontuacao = re.compile(r'[^\w\s]')
  regex_digitos = re.compile(r'\d+')
  regex_espacos = re.compile(r' +')

  questoes_port_sem_pont = sub_codigo(textos, regex_codigo)#Filtro de substituir codigo
  questoes_port_min = minusculo(questoes_port_sem_pont)#Filtro tornar minusculo
  questoes_port_sem_dig = remover(questoes_port_min, regex_digitos)#Filtro de remover digitos
  questoes_port_espaco = remover_espacos(questoes_port_sem_dig, regex_espacos)#Filtro de remover espaços duplicados
  return questoes_port_espaco
```

# Modelo de linguagem

```python
from nltk.util import bigrams
from nltk.lm.preprocessing import pad_both_ends
from nltk.lm.preprocessing import padded_everygram_pipeline
from collections import Counter
from sklearn.model_selection import train_test_split
from nltk.tokenize import WhitespaceTokenizer
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import MLE
from nltk.lm import Laplace
```

```python
overflow_port["idioma"]="port"
overflow_ingles["idioma"]="ing"
overflow_esp["idioma"]="esp"
```

```python
port_treino,port_test = train_test_split(overflow_port.Tratamento_final, test_size=0.2, random_state=42)
```

```python
ing_treino,ing_test = train_test_split(overflow_ingles.Tratamento_final, test_size=0.2, random_state=42)
```

```python
esp_treino,esp_test = train_test_split(overflow_esp.Tratamento_final, test_size=0.2, random_state=42)
```

```python
# Juntas todas as questões em uma string
todas_quest_port = ' '.join(port_treino)
todas_quest_ing = ' '.join(ing_treino)
todas_quest_esp = ' '.join(esp_treino)
```

```python
# Forma o conjunto de todas as palvras conhecidas
todas_palavras_port = WhitespaceTokenizer().tokenize(todas_quest_port)
todas_palavras_ing = WhitespaceTokenizer().tokenize(todas_quest_ing)
todas_palavras_esp = WhitespaceTokenizer().tokenize(todas_quest_esp)
```

```python
# Forma os bigrmas e o vocabualrios das linguas
port_treino_bigram, vocab_port = padded_everygram_pipeline(2, todas_palavras_port)
ing_treino_bigram, vocab_ing = padded_everygram_pipeline(2, todas_palavras_ing)
esp_treino_bigram, vocab_esp = padded_everygram_pipeline(2, todas_palavras_esp)
```

```python
# Definindo o modelo
modelo_port = MLE(2)
modelo_port.fit(port_treino_bigram, vocab_port)
```

```python
# Teste da capcidade do modelo gerar palavras
modelo_port.generate(num_words=10)
```

```python
# Função de treino do modelo
def train_model_mle(lista_texto):
  all_quest = ' '.join(lista_texto)
  all_words = WhitespaceTokenizer().tokenize(all_quest)
  bigrams, vocabulary = padded_everygram_pipeline(2, all_words)
  model = MLE(2)
  model.fit(bigrams, vocabulary)
  return model
```

```python
modelo_port_2 = train_model_mle(port_treino)
modelo_port_2.generate(num_words=10)
```

```python
modelo_ing = train_model_mle(ing_treino)
modelo_ing.generate(num_words=10)
```

```python
modelo_esp = train_model_mle(esp_treino)
modelo_esp.generate(num_words=10)
```

```python
def cal_perplexidade(model, texto):
  perplexidade=0
  words = WhitespaceTokenizer().tokenize(texto)
  words_fakechar = [list(pad_both_ends(word,n=2)) for word in words]
  words_bigramns = [list(bigrams(word)) for word in words_fakechar]
  #print(words_bigramns)
  for word in words_bigramns:
    #print(word)
    perplexidade += model.perplexity(word)
  return perplexidade
```

```python
print(cal_perplexidade(modelo_ing, "good morning"))
```

```python
# A perplexidade deu infinito pois o modelo nunca viu essas palavras
print(cal_perplexidade(modelo_ing, "Bom dia"))
```

Modelo de Laplace

```python
# Função de treino do modelo
def train_model_Laplace(lista_texto):
  all_quest = ' '.join(lista_texto)
  all_words = WhitespaceTokenizer().tokenize(all_quest)
  bigrams, vocabulary = padded_everygram_pipeline(2, all_words)
  model = Laplace(2)
  model.fit(bigrams, vocabulary)
  return model
```

```python
model_ing_Lap = train_model_Laplace(ing_treino)
model_ing_Lap.generate(num_words=10)
```

```python
modelo_port_Lap = train_model_Laplace(port_treino)
modelo_port_Lap.generate(num_words=10)
```

```python
modelo_esp_Lap = train_model_Laplace(esp_treino)
modelo_esp_Lap.generate(num_words=10)
```

```python
# Com o modelo de Laplace não se tem o perplexidade infinita
print(cal_perplexidade(model_ing_Lap, "Bom dia"))
```

```python
print(cal_perplexidade(modelo_port_Lap, "Bom dia"))
```

```python
def idiom_search(lista_textos):
  idiom = []
  for texto in lista_textos:
    port_perp = cal_perplexidade(modelo_port_Lap, texto)
    ing_perp = cal_perplexidade(model_ing_Lap, texto)
    esp_perp = cal_perplexidade(modelo_esp_Lap, texto)
    if ing_perp >= port_perp <= esp_perp:
      idiom.append("port")
    elif port_perp > ing_perp < esp_perp:
      idiom.append("ing")
    else:
      idiom.append("esp")
  return idiom
```

```python
result_port = idiom_search(port_test)
result_ing = idiom_search(ing_test)
result_esp = idiom_search(esp_test)
```

```python
taxa_acerto=result_port.count("port")/len(port_test)
print(f"acuracia portugues: {taxa_acerto}")
```

```python
taxa_acerto=result_ing.count("ing")/len(ing_test)
print(f"acuracia ingles: {taxa_acerto}")
```

```python
taxa_acerto=result_esp.count("esp")/len(esp_test)
print(f"acuracia espanhol: {taxa_acerto}")
```
